```python
df = pd.read_csv("/kaggle/input/predict-the-age-of-a-trilobite/trilobite.csv")
df.shape
```

```python
# All columns with missing values (excluding the one to drop)
columns_to_fill = [
    "state",
    "country",
    "latlng_basis",
    "formation",
    "stratigraphy_scale",
    "lithology",
    "environment",
    "assembly_composition",
    "preservation_mode",
    "collection_type",
    "vision",
    "time_period"
]

# Fill missing values with "Unknown"
df[columns_to_fill] = df[columns_to_fill].fillna("Unknown")

# Drop 'late_interval' due to ~91% missingness
df.drop(columns=["late_interval"], inplace=True)
```
